```
#Calcula nota final dado que AP1 e AP2 tem peso 40% e AP3 20%
def nota_final(AP1,AP2,AP3):
  Peso1 = 4
  Peso2 = 4
  Peso3 = 2
  TotalPesos = Peso1+Peso2+Peso3
  return ((AP1*Peso1) + (AP2*Peso2) + (AP2*Peso3) ) / (TotalPesos)
print(nota_final(8,5,6))
```

```
def nota_conceito(nota):
  if nota >= 9:
    return('A')
  elif nota >=7 and nota <=9:
    return('B')
  elif nota >=5 and nota <=7:
    return('D')
  else: 
    return('F')

print(nota_conceito(5.8))
```
