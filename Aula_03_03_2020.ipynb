CONDIÇÕES IF

```
def aprovacao(nota,freq):
    if (nota>=7 and freq >=0.75):
      return('Aprovado')
    else:
      return('Reprovado')
      
aprov = aprovacao(9,0.9)
print(aprov)
```

LAÇO / LOOP (FOR)

```
def media_turma():
    notas = [5,5,6,6,7,7,8,8,9,9]
    soma = 0
    for i in range (0,10,1):
        soma = soma + notas[i]
    return(soma/len(notas))

print(media_turma())
```

CALCULE A NOTA FINAL, DADO QUE AP1 E AP2 TEM PESO 40% E AP3 20%

```
def nota_final(ap1,ap2,ap3):
  nf = (ap1*0.4) + (ap2*0.4) + (ap3*0.2)
  return(nf)

print(nota_final(8,7,6))
```

CALCULE A NOTA COMO CONCEITO, COM SENDO A(MAIOR QUE 9), B (ENTRE 7 E 9), D(ENTRE 5 E 7), F(ABAIXO DISTO)

```
def nota_conceito(nota):
    if (nota>=9):
      return('A')
    elif ((nota>=7) and (nota<9)):
      return('B')
    elif ((nota>=5) and (nota<7)):
      return('D')
    else:
      return('F')
      
nota = nota_conceito(6)
print(nota)
```

CALCULE A MAIOR NOTA DA TURMA COM NOTAS: 7,6,9,9,8,8,10,5,6 E 7

```
def maior_nota():
    notas = [7,6,9,9,8,8,10,5,6,7]
    maior = max(notas)
    return(maior)

print(maior_nota())
```

```
def maior_nota1():
    notas = [7,6,9,9,8,8,10,5,6,7]
    maior = 0
    
    for i in range (0,10,1):
      if (maior>notas[i]):
        maior = maior
      else:
        maior = notas[i]
    return(maior)

print(maior_nota1())
```
