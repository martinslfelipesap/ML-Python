```
import pandas as pd
treino = pd.read_csv('/content/drive/My Drive/PROJ/dados/volei - ok.csv')
treino
```

```
teste = pd.read_csv('/content/drive/My Drive/PROJ/dados/volei - prd.csv')
teste
```

```
import tensorflow
from tensorflow import keras
rn = keras.Sequential([
  keras.layers.Input(len(treino.columns)-1),
  keras.layers.Dense(3, 'relu'),
  keras.layers.Dense(2, 'relu'),
  keras.layers.Dense(1, 'sigmoid'),
])
rn.compile(optimizer='adam' , loss='binary_crossentropy')
rn.fit(treino[['previsao', 'temperatura', 'umidade', 'vento']], treino['jogar'])
previsoes = rn.predict(teste[['previsao', 'temperatura', 'umidade', 'vento']])
previsoes
```

```
def ativacao(v):
  limiar = 0.05
  if (v > limiar):
    return 1
  else:
    return 0

for previsao in previsoes:
  print(ativacao(previsao))
```

# **Exercicio Rede Neural**

```
import pandas as pd
treino_ex = pd.read_csv('/content/drive/My Drive/PROJ/dados/RiscoCredito - okk.csv')
teste_ex = pd.read_csv('/content/drive/My Drive/PROJ/dados/RiscoCredito - prever.csv')
print(treino_ex)
teste_ex
```

```
#treino_ex = Inadimplente 	Genero 	GrauEscolaridade 	Profissao 	Renda
#teste_ex  = Genero 	GrauEscolaridade 	Profissao 	Renda
import tensorflow
from tensorflow import keras
rn = keras.Sequential([
  keras.layers.Input(len(treino_ex.columns)-1), #quantidade de colunas dos dados de treino -1 (RESULTADO)
  keras.layers.Dense(3, 'relu'), # 3 nós
  keras.layers.Dense(2, 'relu'), # 3 nós
  keras.layers.Dense(2, 'relu'), # 3 nós
  keras.layers.Dense(1, 'sigmoid'), # 1 nó de saída
])

rn.compile(optimizer='adam' , loss='binary_crossentropy')
rn.fit(treino_ex[['Genero', 'GrauEscolaridade', 'Profissao', 'Renda']], treino_ex['Inadimplente']) # dados de treino

previsoes = rn.predict(teste) # dados de teste
previsoes
```

```
def ativacao(v):
  limiar = 0.53
  if (v > limiar):
    return 1
  else:
    return 0

for previsao in previsoes:
  print(ativacao(previsao))
```
