```
import pandas as pd
import seaborn as sns
#from sklearn.linear_model import LinearRegression
#from sklearn.neighbors import KNeighborsClassifier
#from sklearn.naive_bayes import MultinomialNB
```

```
sal_poup = pd.read_csv('https://raw.githubusercontent.com/lcbjuk/ML/master/dados/salario_conta%20-%20ok.csv')
sal_poup
```

```
from sklearn.cluster import KMeans

k = KMeans(n_clusters=5)
k.fit(sal_poup)
```

```
import seaborn as sns
sns.scatterplot(sal_poup['salario'], sal_poup['saldo'], hue=k.labels_, palette=sns.color_palette('Set1',5))
```

**Exercício KMeans para jogar volei**



```

volei = pd.read_csv('/content/drive/My Drive/PROJ/dados/volei - ok.csv')
volei
```

```
#  	previsao 	temperatura 	umidade 	vento 	jogar
k = KMeans(n_clusters=2) #numero de opçoes, neste caso jogar e nao jogar
k.fit(volei) #entrada dataframe
sns.scatterplot(volei['previsao'], volei['temperatura'], hue=k.labels_)
```
